```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

```python
credit_card = pd.read_csv('/content/drive/MyDrive/Codsoft Dataset/creditcard.csv')
credit_card.head(15)
```

```python
#checking fraud count, percentage
fraud_count = credit_card['Class'].value_counts()
fraud_percent = 100*fraud_count/credit_card.shape[0]
fraud_data = pd.concat([fraud_count, fraud_percent], axis = 1).reset_index()
fraud_data.columns = ['Class','Count','Percentage']

fraud_data
```

```python
# Handling imbalance
credit_card_fraud = credit_card[credit_card['Class']==1]
credit_card_not_fraud = credit_card[credit_card['Class']==0]
credit_card_not_fraud_sampled = credit_card_not_fraud.sample(n=len(credit_card_fraud), replace=False, random_state = 101)

credit_card_balanced = pd.concat([credit_card_not_fraud_sampled, credit_card_fraud], axis=0).sample(frac=1, replace=False, random_state=101).reset_index().drop('index',axis = 1)

credit_card_balanced
```

```python
# Checking the balanced data Count and Perentage
fraud_count = credit_card_balanced['Class'].value_counts()
fraud_percent = 100*fraud_count/credit_card_balanced.shape[0]
fraud_data = pd.concat([fraud_count, fraud_percent], axis = 1).reset_index()
fraud_data.columns = ['Class','Count','Percentage']

fraud_data
```

```python
# train and test
X_train, X_test, y_train, y_test = train_test_split(credit_card_balanced.drop('Class', axis=1), credit_card_balanced['Class'], test_size=0.2, random_state=101)

print(f''' X_train: {X_train.shape}
      X_test: {X_test.shape}
      y_train: {y_train.shape}
      y_test: {y_test.shape}
      ''')
```

```python
# Logistic model Pipeline

RandomForestModel= Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier())
])

RandomForestModel.fit(X_train, y_train)
```

```python
# Prediction

y_predict = RandomForestModel.predict(X_test)
y_predict
```

```python
# accuracy Classification report
cr = classification_report(y_test, y_predict)
print(cr)
```

```python

# Save the model
pickle.dump(RandomForestModel, open('RandomForestModel.pkl', 'wb'))
```
